```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey_results_public.csv")
```

```python
df.head()
```

```python
#data cleaning


#1 to keep only important columns
df = df[["Country","EdLevel","YearsCodePro","Employment","ConvertedComp"]]
df = df.rename({"ConvertedComp":"Salary"},axis = 1)#  'columns' can also be written in place of 1
df.head()
```

```python
#2 now we only want to keep those people with some salary ie. we'll  be removing all those with salary NaN
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
#3 now we'll  be dropping all data points with NaN
df = df.dropna()# we can also place mean in place of NaN
df.isnull().sum() # to check if any null value is left
```

```python
df =df[df["Employment"] == "Employed full-time"]
df = df.drop("Employment",axis = 'columns')
df.head()
```

```python
df["Country"].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else: 
            categorical_map[categories.index[i]] = 'other'
        return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(),400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```
